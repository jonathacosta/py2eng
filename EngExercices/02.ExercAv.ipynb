### Solução 01 - programação manual

```python
import numpy as np
import matplotlib.pyplot as plt

# Funçoes
def f(x):    
    return x**2 + 3*x - 2
def g(x):
    return 3*x**4 - 2*x**2 + 3*x**2 - 14
def h(x):
    return 5*x**3 - 2*x**2 + 3*x + 1
def df(x):
    return 2*x* + 3
def dg(x):
    return 12*x**3 - 4*x + 6*x
def dh(x):
    return 15*x**2 - 4*x + 3

# Graficos
x=np.linspace(0,100)
plt.plot(x,f(x),label="f(x)")
plt.plot(x,g(x),label="g(x)")
plt.plot(x,h(x),label="h(x)")
plt.plot(x,df(x),label="df/dx")
plt.plot(x,dg(x),label="dg/dx")
plt.plot(x,dh(x),label="dh/dx")
plt.grid(True)
plt.legend()
plt.show()
```

### Solução 05 - Utilizando o sympy

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

#***************************************************
# Funções e derivadas
def f(x):
    return x**2 + 3*x - 2

def df(f): 
    #Definindo manualmente a derivada primeira para cada função
    x = sp.Symbol('x')
    fun = f(x)    
    df = sp.diff(fun,x,1)
    return sp.lambdify('x',df,modules='numpy')                  

def g(x):
    return 3*x**4 - 2*x**2 + 3*x**2 - 14

def dg(g): 
    x = sp.Symbol('x')
    fun = g(x)    
    dg=sp.diff(fun,x,1)
    dg1=sp.lambdify(x,dg,'numpy') 
    return dg1    

def h(x):
    return 5*x**3 - 2*x**2 + 3*x + 1

def dh(h): 
    x = sp.Symbol('x')
    fun = h(x)    
    dh=sp.diff(fun,x,1)
    dh1=sp.lambdify(x,dh,'numpy') 
    return dh1    

#***************************************************
def graphs(f,g,h,df,dg,dh):
    x=np.linspace(0,100)
    plt.plot(x,f(x),label="f(x)")
    plt.plot(x,g(x),label="g(x)")
    plt.plot(x,h(x),label="h(x)")
    plt.plot(x,df(x),label="df/dx")
    plt.plot(x,dg(x),label="dg/dx")
    plt.plot(x,dh(x),label="dh/dx")
    plt.grid(True)
    plt.legend()
    plt.show()

df,dg,dh = df(f), dg(g), dh(h)    
graphs(f,g,h,df,df,dh)
```
