## **Estruturas de Dados em Python**
- são formas de organizar e armazenar dados. Python oferece várias estruturas de dados embutidas.
- Duas das estruturas de dados mais fundamentais são **Listas e Tuplas** (estruturas de dados ordenadas) e **Dicionários e Conjuntos** (estruturas de dados não ordenadas).

### Listas e Tuplas:

#### 1. **Listas:**
Uma lista é uma coleção ordenada e mutável de itens. Você pode armazenar qualquer tipo de dado em uma lista e ela é definida usando colchetes `[]`.

```python
lista = [1, 2, 3, 4, 5]

# Índices e Fatiamento
print(lista[0])  # Saída: 1 (índice começa do zero)
print(lista[1:4])  # Saída: [2, 3, 4] (fatiamento)

# Métodos de Lista
lista.append(6)  # Adiciona 6 ao final da lista
lista.pop(2)     # Remove o elemento no índice 2, valor = 3
del lista[2]     # Remove o elemento no índice 2, valor = 4 após a operação acima.
```

#### 2. **Tuplas:**
Uma tupla é uma coleção ordenada e imutável de itens. É definida usando parênteses `()`.

```python
tupla = (1, 2, 3, 4, 5)

# Índices e Fatiamento (semelhante às listas)
print(tupla[0])  # Saída: 1
print(tupla[1:4])  # Saída: (2, 3, 4)

# As tuplas são imutáveis, então não têm métodos de alteração como as listas
```

### Dicionários e Conjuntos:

#### 3. **Dicionários:**
Um dicionário é uma coleção não ordenada de dados em pares chave-valor. É definido usando chaves `{}`.

```python
pessoa = {
    "nome": "Alice",
    "idade": 30,
    "profissao": "Engenheira"
}

# Acesso por Chave
print(pessoa["nome"])  # Saída: Alice

# Adição e Atualização de Pares Chave-Valor
pessoa["cidade"] = "Nova York"
pessoa["profissao"] = "Cientista de Dados"
```

#### 4. **Conjuntos:**
Um conjunto é uma coleção não ordenada e mutável de itens únicos. É definido usando chaves `{}` ou a função `set()`.

```python
conjunto1 = {1, 2, 3, 4, 5}
conjunto2 = {3, 4, 5, 6, 7}

# Operações de Conjunto
uniao = conjunto1.union(conjunto2)  # União de conjuntos
intersecao = conjunto1.intersection(conjunto2)  # Interseção de conjuntos
```

**Dicas Adicionais:**
- As listas e os dicionários podem conter itens de diferentes tipos de dados.
- As tuplas e os conjuntos são imutáveis (tuplas) ou mutáveis (conjuntos), mas não podem conter itens de diferentes tipos de dados.

Essas estruturas de dados são essenciais para a manipulação eficiente de dados em Python e são amplamente usadas em várias aplicações. Conhecê-las e compreendê-las é fundamental para a programação eficaz em Python.
