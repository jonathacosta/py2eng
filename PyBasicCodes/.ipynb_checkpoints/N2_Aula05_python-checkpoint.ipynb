## **Projetos propostos em POO:**

**Engenharia de Controle e Automação:**

1. Crie uma classe `ControladorPID` que represente um controlador PID com métodos para calcular a saída do sistema.
2. Implemente uma classe `SensorTemperatura` que simule um sensor de temperatura, fornecendo métodos para ler e converter valores de temperatura.
3. Crie uma classe `ControladorPI` que represente um controlador PI com métodos para calcular a saída do sistema.
4. Implemente uma classe `ControladorPD` que represente um controlador PD com métodos para calcular a saída do sistema.

**Engenharia Mecânica:**

1. Implemente uma classe `MotorEletrico` que simule o funcionamento de um motor elétrico, permitindo ligar, desligar e ajustar a velocidade.
2. Crie uma classe `SistemaMolaAmortecedor` que modele um sistema massa-mola-amortecedor, permitindo calcular a resposta do sistema a diferentes condições iniciais.
3. Implemente uma classe `SistemaPoleias` que modele um sistema de polias, permitindo calcular as velocidades e forças nas diferentes partes do sistema.
4. Crie uma classe `AnáliseTensão` que permita calcular tensões em diferentes pontos de uma estrutura com base em forças aplicadas e áreas de seção transversal.

**Matemática:**

1. Desenvolva uma classe `CalculadoraMatematica` que contenha métodos para calcular operações matemáticas como adição, subtração, multiplicação e divisão.
2. Desenvolva uma classe `CalculadoraGeometrica` que contenha métodos para calcular áreas e volumes de diferentes formas geométricas como círculos, retângulos e esferas.
3. Desenvolva uma classe `CalculadoraEstatistica` que contenha métodos para calcular média, mediana e desvio padrão de um conjunto de números.

**Física:**

1. Crie uma classe `MovimentoUniforme` para calcular a posição de um objeto em movimento uniforme com base no tempo, velocidade inicial e aceleração.
2. Implemente uma classe `LeiOhm` que represente a Lei de Ohm, permitindo calcular a resistência elétrica com base na tensão e corrente.
3. Crie uma classe `MovimentoUniformementeAcelerado` para calcular a posição de um objeto em movimento uniformemente acelerado com base no tempo, velocidade inicial e aceleração.
4. Desenvolva uma classe `LeiGravitacaoUniversal` que permita calcular a força gravitacional entre duas massas com base na Lei da Gravitação Universal de Newton.

Esses exercícios oferecem uma variedade de desafios em diferentes áreas da engenharia e ciências exatas, fornecendo uma ampla prática em programação orientada a objetos.
