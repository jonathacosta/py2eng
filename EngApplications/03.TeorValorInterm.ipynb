* Perceba que este teorema diz que $\exists$ pelo menos um número $c$, mas não fornece um método para encontrar seu valor.
   
* É chamado de Teorema da Existência e implica que o gráfico de uma função contínua não pode apresentar um salto vertical.

### Classes sem o método init

```python
import numpy as np
import matplotlib.pyplot as plt

'''
Na estrutura de classe, os métodos são concatenados num único estrutura que o correlacina - a classe -, e as entradas principais ocorrem nos métodos contidos na classe, e as chamadas ocorrem a partir o objetivo instanciado à classe.
"Nome_do_objeto = Nome_da_classe.Metodo(atributos_do_método)"
k=JRC().graf(1,10,10)

''' 

class JRC:    
    def f(self,x):
        return x**4 + 2.*x+4.

    def al(self,a,b):
        import random as rd
        if (b-a)<=1.:
            return a+rd.random() 
        else:
            return  rd.randrange(int(a),int(b))+rd.random()
           
    def graf(self,a,b,n=1):
        '''
        Informe a faixa de domínio (a,b) e a quantidade de pontos aleatórios (n).
        '''
        plt.figure(figsize=(4,3))
        x0= np.linspace(a,b)
        y0= self.f(x0)  
        plt.plot(x0,y0,label='f(x)')
        x1= np.array([ self.al(a,b) for i in range(int(n)) ])
        y1= self.f(x1)
        plt.plot(x1,y1,'g*',label='x aleatório')
        plt.title("Teorema do valor intermediário")
        plt.legend()
        plt.style.use('ggplot')
```

```python
# Um mesmo objeto com atributos diferentes
s=JRC()
s.graf(1,10,1)
s.graf(1,2,3)
```

### Classe com o método init

```python
'''
Neste método, as entradas principais são na classe.
Entradas adicionais, nas funções; se necessário.   
Alerta para não carregar com atribuitos extras. 
Se a função é definida como al(self), a chamada dela deve ser vazia al(self)
'''  

class JRC:    
    def __init__(self,a,b,n):
        self.a=a
        self.b=b
        self.n=n
    
    def f(self,x):
        return x**4 + 2.*x+4.

    def al(self):
        import random as rd
        if (self.b-self.a)<=1.:
            return self.a+rd.random() 
        else:
            return  rd.randrange(int(self.a),int(self.b))+rd.random()
           
    def graf(self): 
        plt.figure(figsize=(4,3))
        x0= np.linspace(self.a,self.b)
        y0= self.f(x0)  
        plt.plot(x0,y0,label='f(x)')
        x1= np.array([ self.al() for i in range(int(self.n)) ])
        y1= self.f(x1)
        plt.plot(x1,y1,'g*',label='x aleatório')
        plt.title(f'Teorema do valor intermediário com {self.n} pontos')
        plt.legend()
        plt.style.use('ggplot')
```

```python
#Objetos de uma mesma classe com atributos distintos
l=JRC(1,10,5).graf()
m=JRC(1,2,3).graf()
```
